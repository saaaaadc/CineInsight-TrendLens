```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (download from Kaggle: https://www.kaggle.com/datasets/shivamb/netflix-shows)
df = pd.read_csv('netflix_movies_detailed_up_to_2025.csv')

# Display basic info
print("\nDataset Overview:")
df.info()
```

```python
indian_shows = df[df['country'].str.contains('India', na=False)]
indian_shows
```

```python
# Filter for South Indian movies based on common South Indian languages
south_indian_languages = ['ta', 'te', 'ml', 'kn']  # Tamil, Telugu, Malayalam, Kannada
south_indian_movies = indian_shows[indian_shows['language'].isin(south_indian_languages)]
south_indian_movies
```

```python
plt.figure(figsize=(10, 6))
indian_shows['release_year'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Number of Indian Movies Released by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
```

```python
# Prepare dashboard_df: count Indian and Global movies by release year
indian_counts = indian_shows['release_year'].value_counts().sort_index()
global_counts = df['release_year'].value_counts().sort_index()

dashboard_df = pd.DataFrame({
	'Indian Movies': indian_counts,
	'Global Movies': global_counts
}).fillna(0).astype(int)

# Plot Indian Movies Released by Year
plt.figure(figsize=(10, 5))
dashboard_df['Indian Movies'].plot(kind='bar', color='orange')
plt.title('Number of Indian Movies Released by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()

# Plot Global Movies Released by Year
plt.figure(figsize=(10, 5))
dashboard_df['Global Movies'].plot(kind='bar', color='steelblue')
plt.title('Number of Global Movies Released by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.tight_layout()
plt.show()
```

```python
# Extract year and genres for global movies
global_movies = df[df['country'].notna()]  # Filter out rows with missing country
global_movies = global_movies.explode('genres')  # If genres is a list, otherwise split by comma
if global_movies['genres'].apply(lambda x: isinstance(x, str)).all():
    global_movies = global_movies.assign(genres=global_movies['genres'].str.split(','))
global_movies = global_movies.explode('genres')
global_movies['genres'] = global_movies['genres'].str.strip()

# Count plot of top genres over time
top_genres = global_movies['genres'].value_counts().head(10).index
filtered = global_movies[global_movies['genres'].isin(top_genres)]

plt.figure(figsize=(14, 7))
sns.countplot(data=filtered, x='release_year', hue='genres', order=sorted(filtered['release_year'].unique()))
plt.title('Top Global Movie Genres Over Time')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Prepare Indian movies genres for count plot
indian_genres = indian_shows.copy()
if indian_genres['genres'].apply(lambda x: isinstance(x, str)).all():
    indian_genres = indian_genres.assign(genres=indian_genres['genres'].str.split(','))
indian_genres = indian_genres.explode('genres')
indian_genres['genres'] = indian_genres['genres'].str.strip()

# Get top 10 genres for Indian movies
top_indian_genres = indian_genres['genres'].value_counts().head(10).index
filtered_indian = indian_genres[indian_genres['genres'].isin(top_indian_genres)]

plt.figure(figsize=(14, 7))
sns.countplot(data=filtered_indian, x='release_year', hue='genres', order=sorted(filtered_indian['release_year'].unique()))
plt.title('Top Indian Movie Genres Over Time')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Calculate average IMDb (vote_average) rating by year for Indian and Global movies
indian_rating_trend = indian_shows.groupby('release_year')['vote_average'].mean()
global_rating_trend = global_movies.groupby('release_year')['vote_average'].mean()

plt.figure(figsize=(12, 6))
plt.plot(indian_rating_trend.index, indian_rating_trend.values, marker='o', label='Indian Movies')
plt.plot(global_rating_trend.index, global_rating_trend.values, marker='o', label='Global Movies')
plt.title('IMDb Rating Trends: Indian vs Global Movies')
plt.xlabel('Release Year')
plt.ylabel('Average IMDb Rating')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Predict future trending genres by extrapolating genre counts and visualizing the trend

# Prepare data: count top genres by year for global movies
genre_year_counts = global_movies.copy()
if genre_year_counts['genres'].apply(lambda x: isinstance(x, str)).all():
    genre_year_counts = genre_year_counts.assign(genres=genre_year_counts['genres'].str.split(','))
genre_year_counts = genre_year_counts.explode('genres')
genre_year_counts['genres'] = genre_year_counts['genres'].str.strip()

# Focus on top 5 genres overall
top5_genres = genre_year_counts['genres'].value_counts().head(5).index
filtered = genre_year_counts[genre_year_counts['genres'].isin(top5_genres)]

# Group by year and genre
genre_trend = filtered.groupby(['release_year', 'genres']).size().unstack(fill_value=0)

# Extrapolate next 3 years using linear trend
future_years = [2026, 2027, 2028]
trend_pred = {}

for genre in top5_genres:
    y = genre_trend[genre].values
    X = genre_trend.index.values.reshape(-1, 1)
    model = LinearRegression().fit(X, y)
    y_future = model.predict(np.array(future_years).reshape(-1, 1))
    trend_pred[genre] = y_future

# Plot historical and predicted genre trends
plt.figure(figsize=(14, 7))
for genre in top5_genres:
    plt.plot(genre_trend.index, genre_trend[genre], marker='o', label=f"{genre} (history)")
    plt.plot(future_years, trend_pred[genre], marker='x', linestyle='--', label=f"{genre} (predicted)")

plt.title('Trending Movie Genres: Historical and Predicted')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from textblob import TextBlob

# Compute sentiment polarity for each description if not already present
if 'sentiment' not in df.columns:
    df['sentiment'] = df['description'].dropna().apply(lambda x: TextBlob(x).sentiment.polarity)
    df['sentiment'] = df['sentiment'].fillna(0)
else:
    df['sentiment'] = df['sentiment'].fillna(0)

# Prepare genres for analysis
genre_sentiment_df = df[['genres', 'sentiment']].dropna(subset=['genres'])
genre_sentiment_df = genre_sentiment_df.assign(genres=genre_sentiment_df['genres'].str.split(','))
genre_sentiment_df = genre_sentiment_df.explode('genres')
genre_sentiment_df['genres'] = genre_sentiment_df['genres'].str.strip()

# Get top genres for visualization
top_genres_for_sentiment = genre_sentiment_df['genres'].value_counts().head(10).index
filtered_sentiment = genre_sentiment_df[genre_sentiment_df['genres'].isin(top_genres_for_sentiment)]

# Calculate average sentiment per genre
avg_sentiment = filtered_sentiment.groupby('genres')['sentiment'].mean().sort_values()

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_sentiment.values, y=avg_sentiment.index, palette='coolwarm')
plt.title('Average Sentiment Polarity by Genre')
plt.xlabel('Average Sentiment Polarity')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python
# Use genre_sentiment_df and create sentiment_label
sentiment_df = genre_sentiment_df.copy()
sentiment_df['sentiment_label'] = sentiment_df['sentiment'].apply(
	lambda x: 'Positive' if x > 0.1 else ('Negative' if x < -0.1 else 'Neutral')
)

plt.figure(figsize=(12, 6))
sns.countplot(
	data=sentiment_df,
	x='sentiment_label',
	hue='genres',
	order=['Positive', 'Neutral', 'Negative'],
	palette='coolwarm'
)
plt.title('Sentiment Distribution by Genre')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import pandas as pd

# Assuming 'df' is your loaded DataFrame with Netflix movie data

def get_movie_details(title):
    # Case-insensitive search for the movie title
    result = df[df['title'].str.lower() == title.lower()]
    if not result.empty:
        return result.iloc[0]
    else:
        return None

# Create input text box for movie title
movie_input = widgets.Text(
    value='',
    placeholder='Enter movie title',
    description='🎬 Movie:',
    layout=widgets.Layout(width='350px'),
    style={'description_width': 'initial'}
)

# Create submit button
submit_button = widgets.Button(
    description='🔍 Get Details',
    button_style='success',
    layout=widgets.Layout(width='150px', margin='0 0 0 10px'),
    tooltip='Click to search for movie details'
)

# Output widget to display movie details or messages
output = widgets.Output(layout={'border': '1px solid #4CAF50', 'padding': '10px', 'margin': '10px 0'})

def on_submit(b):
    with output:
        clear_output()
        title = movie_input.value.strip()
        if not title:
            print("Please enter a movie title.")
            return
        details = get_movie_details(title)
        if details is not None:
            # Helper function to safely display values or 'N/A' if missing
            def safe(val):
                return val if pd.notnull(val) and val != '' else "<span style='color:#888;'>N/A</span>"
            
            display(widgets.HTML(
                f"""
                <div style="font-family:Arial; font-size:16px; line-height:1.7;">
                    <b>🎬 Title:</b> {safe(details.get('title'))}<br>
                    <b>📅 Year:</b> {safe(details.get('release_year'))}<br>
                    <b>⭐ IMDb Rating:</b> {safe(details.get('vote_average'))}<br>
                    <b>🎭 Genres:</b> {safe(details.get('genres'))}<br>
                    <b>🌍 Country:</b> {safe(details.get('country'))}<br>
                    <b>🎤 Cast:</b> {safe(details.get('cast'))}<br>
                    <b>🎬 Director:</b> {safe(details.get('director'))}<br>
                    <b>📝 Description:</b> {safe(details.get('description'))}<br>
                    <b>💰 Budget:</b> {safe(details.get('budget'))}<br>
                    <b>💵 Revenue:</b> {safe(details.get('revenue'))}<br>
                    <b>🎫 Popularity:</b> {safe(details.get('popularity'))}<br>
                    <b>🗳️ Vote Count:</b> {safe(details.get('vote_count'))}<br>
                    <b>💬 Sentiment:</b> {safe(details.get('sentiment'))}<br>
                    <b>🕒 Date Added:</b> {safe(details.get('date_added'))}<br>
                    <b>⏱️ Duration:</b> {safe(details.get('duration'))}<br>
                    <b>🔤 Language:</b> {safe(details.get('language'))}<br>
                    <b>🔖 Rating:</b> {safe(details.get('rating'))}<br>
                </div>
                """
            ))
        else:
            display(widgets.HTML(
                "<span style='color:red; font-weight:bold;'>Movie not found. Please check the title and try again.</span>"
            ))

# Bind the button click event to the handler
submit_button.on_click(on_submit)

# Layout the input box and button horizontally
ui_box = widgets.HBox([movie_input, submit_button])

# Display the full UI vertically: title, input+button, output area
display(widgets.VBox([
    widgets.HTML("<h3 style='color:#388e3c; font-family:Arial;'>Netflix Movie Explorer</h3>"),
    ui_box,
    output
]))
```
